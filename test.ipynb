```python
from dotenv import load_dotenv
load_dotenv()
import os
import pandas as pd
import pickle
from amazon.paapi import AmazonAPI
amazon = AmazonAPI(os.getenv("KEY"), os.getenv("SECRET"), os.getenv("STORE"), "US")
```

```python
products = pd.read_csv('products2.csv')
productsCodeToNameMap = {}
productsCodeToExamplesMap = {}

for index, row in products.iterrows():
    if row['Include'] == "Yes":
        productsCodeToNameMap[row['Code']] = row['Name']
        
        examples = []
        if not pd.isnull(row['Example 1']):
            examples += [row['Example 1']]
        if not pd.isnull(row['Example 2']):
            examples += [row['Example 2']]
        if not pd.isnull(row['Example 3']):
            examples += [row['Example 3']]
        if (len(examples) > 0):
            productsCodeToExamplesMap[row['Code']] = examples
```

```python
for product, examples in productsCodeToExamplesMap.items():
    items_count = 10
    if len(examples) == 2:
        items_count = 15
    if len(examples) == 1:
        items_count = 30
    
    amazon_products = []
    for example in examples:
        print(product, items_count, example)
        amazon_products += amazon.search_products(item_count=items_count, keywords=example)
    pickle.dump( amazon_products, open( "pickles/" + product  + ".p", "wb" ) )

    
```

```python
from os import listdir
from os.path import isfile, join
all_pickle_paths = [f for f in listdir("pickles") if isfile(join("pickles", f))]
epa_category_to_amazon_group = {}
distinct_amazon_categories = []
for pickle_path in all_pickle_paths:
    amazon_products = pickle.load(open("pickles/" + pickle_path, "rb"))
    epa_category = pickle_path.split(".")[0]
    epa_category_to_amazon_group[productsCodeToNameMap[epa_category]] = []
    for amazon_product in amazon_products:
        if not amazon_product.info.product_group in epa_category_to_amazon_group[productsCodeToNameMap[epa_category]]:
            epa_category_to_amazon_group[productsCodeToNameMap[epa_category]] += [amazon_product.info.product_group]
        if not amazon_product.info.product_group in distinct_amazon_categories:
            distinct_amazon_categories += [amazon_product.info.product_group]
```
